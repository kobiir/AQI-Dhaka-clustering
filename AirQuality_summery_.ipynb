```python
import pandas as pd

FILE_NAME = "air_quality_index_dataset.xlsx"
excel_file = pd.ExcelFile(FILE_NAME)

city_summaries = {}

for city in excel_file.sheet_names:
    df = pd.read_excel(FILE_NAME, sheet_name=city)
    numeric_cols = df.select_dtypes(include='number')

    summary = pd.DataFrame({
        "count": numeric_cols.count(),
        "mean": numeric_cols.mean(),
        "median": numeric_cols.median(),
        "variance": numeric_cols.var(),
        "std": numeric_cols.std(),
        "min": numeric_cols.min(),
        "max": numeric_cols.max()
    })

    city_summaries[city] = summary

with pd.ExcelWriter("citywise_summary.xlsx") as writer:
    for city, summary in city_summaries.items():
        summary.to_excel(writer, sheet_name=city[:30])

print("✅ 16 city summary file created successfully!")
```

```python
cities = ['Sangsad ', 'BARC', 'Darussalam', 'Gazipur', 'Narayonganj',
          'TV st-Chittagong', 'Agrabad Chittagong', 'Sylhet', 'Khulna',
          'Rajshahi', 'Barishal', 'Savar', 'Rangpur', 'Mymansingh',
          'Cumilla', 'Narsingdi']

for city in cities:
    print(f"Summary for {city}:")
    display(city_summaries[city])
    print("\n" + "="*60 + "\n")
```
